```python
import pandas as pd
from collections import Counter
import re
import os
import spacy
import tokenize_to_counter
nlp = spacy.load("en_core_web_lg")
```

```python
TEST_PATH_FOR_FILE = "../data/test/1.txt"
TEST_PATH_FOR_DIR = "../data/test/"
```

```python
def extract_and_save(DIR_PATH):
    files = []
    for file in os.listdir(DIR_PATH):
        x = ""
        with open(os.path.join(DIR_PATH + file)) as f:
            for line in f.readlines():
                x += line
        files.append(x)
    extracted_data = pd.DataFrame()
    extracted_data["docs"] = files
    return extracted_data
```

```python
df = extract_and_save(TEST_PATH_FOR_DIR)
```

```python
def map_dataframe_to_counter(df):
    df["docs"] = df["docs"].apply(lambda x : tokenize_to_counter.solve(x))
    return df
```

```python
df_2 = map_dataframe_to_counter(df)
```
